# **`Data Preprocessing`**




drop Unnamed: 0	 column

```python
df.drop("Unnamed: 0", axis =1 , inplace =True)
```

```python
df
```

```python
df['total_spend']=df['TV']+df['Radio']+df['Newspaper']
```

```python
df
```

```python
sns.scatterplot(data=df , x='total_spend',y='Sales')
```

```python
 sns.regplot(data=df , x='total_spend',y='Sales')
```

split data into feature and target

1- features ---> total_spend(tv , radio and newspaper)---> x

2- target(value of predict) ---> sales---> y

```python
x = df.drop(['Sales','total_spend'],axis=1)
y= df['Sales']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42, shuffle=True)
```

```python
x_train
```

```python
print("x_train shape" , x_train.shape)
```

```python
print("x_test shape" , x_test.shape)
```

```python
print("y_train shape" , y_train.shape)
```

```python
print("x_test shape" , x_test.shape)
```

# **4-Model Building**

```python
#linear regression model
from sklearn.linear_model import LinearRegression
```

model ---> fit(train) ---> predict(test)

machine learning--->fit/predict

```python
#call linear regression model
LR_model = LinearRegression()
```

```python
#fit model on the data (model training)---> (x_train , y_train) /train
LR_model.fit(x_train , y_train)
```

```python
#predict on the test data /test
y_pred = LR_model.predict(x_test)
```
